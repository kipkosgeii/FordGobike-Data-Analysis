```python
import folium
```

```python
# basic 
map_nyc = folium.Map(location=[40.7128, -74.0060], zoom_start=6,tiles='CartoDB Positron')

map_nyc_marker = folium.Marker(location=[40.7128,-74.0060], popup='<strong>Times Square</strong>', tooltip='Click for more info')

map_nyc_marker.add_to(map_nyc)

map_nyc
```

```python
import csv
import folium

map_data = folium.Map(location=[40.7128, -74.0060], zoom_start=12)

with open('data.csv', 'r') as csvfile:

    csv_reader = csv.reader(csvfile)

    for row in csv_reader:
        latitude, longitude, attribute = float(row[0]), float(row[1]),row[2]

        marker = folium.Marker(location=[latitude,longitude])

        marker.add_to(map_data)


map_data
```
